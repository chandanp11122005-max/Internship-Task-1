```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv',sep='\t')

df.info()
df.describe()
df.isnull().sum()


#Percenatage of Null values in the data set
(df.isnull().sum() / len(df)) * 100


#removing the null value rows from the dataset
df = df.dropna(subset=['Income'])

# Checkcing if null values of income are removed.
df.isnull().sum()


#Check for the duplicates
df.duplicated().sum()


#Drop the duplicates if present
df = df.drop_duplicates()


#Converting the date to coninstent format
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'],dayfirst=True)

text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    print(f"\nColumn: {col}")
    print(df[col].unique())

# Standardaize the values
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].str.lower().str.strip()

#Check again for standardize the values
for col in text_cols:
    print(f"\nColumn: {col}")
    print(df[col].unique())

#Before Renaming the Column Header
df.columns

#Renaming the column header to lowercase version
df.columns = df.columns.str.lower()

df.columns

df.info()

binary_cols = ['acceptedcmp1','acceptedcmp2','acceptedcmp3',
               'acceptedcmp4','acceptedcmp5','complain','response']

for col in binary_cols:
    print(col, df[col].unique())

df[binary_cols] = df[binary_cols].astype('bool')

df.info()

df.to_csv('marketing_campaign_cleaned.csv', index=False)
```
